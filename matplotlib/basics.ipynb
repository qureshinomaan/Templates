# Matplotlib tutorials. 
Matplotlib is an useful tool. I hope to use it a lot in future. So let's learn in it in a little depth.

```python
from matplotlib import pyplot as plt
import random 
import numpy as np
import matplotlib as mpl

%matplotlib inline 
```

```python
dev_x = [x for x in range(25, 36)]
dev_y = [random.randint(1000, 1200) for _ in range(25, 36)]
dev_y2 = [random.randint(1000, 1200) for _ in range(25, 36)]

print(plt.style.available) # see the styles available for plots 
# plt.xkcd() #xkcd comic style 
plt.style.use('ggplot') #Using this style 

plt.plot(dev_x, dev_y, color='#444444', linestyle='--', marker='.', label= 'Python')
plt.plot(dev_x, dev_y2, color='#5a7d9a',label= 'AllDevs', linewidth=3)
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.legend()
plt.title('Median Salary(USD) by age ')

plt.grid(True) 
plt.tight_layout()  # for padding across lines. 

plt.savefig('plot.png')
plt.show()
mpl.rcParams.update(mpl.rcParamsDefault)

```

# Tututorial 2 : Bar Charts 

```python
dev_x = [x for x in range(25, 36)]
dev_y = [random.randint(1000, 1200) for _ in range(25, 36)]
dev_y2 = [random.randint(1000, 1200) for _ in range(25, 36)]
plt.style.use('ggplot')
x_indexes = np.arange(len(dev_x))*3
width = 1   # barwidth

plt.bar(x_indexes - width, dev_y, label= 'Python', width = width)
plt.bar(x_indexes, dev_y2, label= 'Python', width=width)



plt.xlabel('Ages')
plt.ylabel('Salary')
plt.legend()
plt.title('Median Salary(USD) by age ')

plt.savefig('plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
minutes = [x for x in range(10)]
player1 = [random.randint(1, 4) for x in range(10)]
player2 = [random.randint(1, 4) for x in range(10)]
player3 = [random.randint(1, 4) for x in range(10)]

labels = ['player1', 'player2', 'player3']

plt.stackplot(minutes, player1, player2, player3, labels=labels)

plt.legend(loc="upper left")

```

```python
plt.style.use('seaborn')
x = [random.randint(0, 10) for _ in range(10)]
y = [random.randint(0, 10) for _ in range(10)]
colors = [random.randint(0, 10) for _ in range(10)]
sizes = [random.randint(100, 500) for _ in range(10)]

plt.scatter(x, y, s= sizes, marker='o', cmap='Greens', edgecolor='black', linewidth=1, alpha=1, c= colors)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

## Last tutorial : Subplots 

```python
dev_x = [x for x in range(25, 36)]
dev_y = [random.randint(1000, 1200) for _ in range(25, 36)]
dev_y2 = [random.randint(1000, 1200) for _ in range(25, 36)]


fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)


print(plt.style.available) # see the styles available for plots 
# plt.xkcd() #xkcd comic style 
plt.style.use('ggplot') #Using this style 

ax1.plot(dev_x, dev_y, color='#444444', linestyle='--', marker='.', label= 'Python')
ax2.plot(dev_x, dev_y2, color='#5a7d9a',label= 'AllDevs', linewidth=3)
ax1.set_xlabel('Ages')
ax1.set_ylabel('Salary')
ax1.legend()
ax1.set_title('Median Salary(USD) by age ')

ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary')
ax2.legend()
ax2.set_title('Median Salary(USD) by age ')



plt.grid(True) 
plt.tight_layout()  # for padding across lines. 

plt.savefig('plot.png')
plt.show()
```
